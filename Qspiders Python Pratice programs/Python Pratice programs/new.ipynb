```python
#WAP to print longest word in given string?
s = input("Enter a string: ")
words = s.split()
longest_word = ""
for word in words:
    if len(word) > len(longest_word):
        longest_word = word
print("Longest word:", longest_word)
```

```python
#WAP to print most repeated character?
s = input("Enter a string: ")
char_count = {}
for char in s:
    char_count[char] = char_count.get(char, 0) + 1
if char_count:
    most_repeated_char = max(char_count, key=char_count.get)
    print("Most repeated character:", most_repeated_char)
else:
    print("Please enter a non-empty string")
```

```python
#WAP to print most repeated word?
def most_repeated_word(sentence):
    # Normalize the string and split into words
    words = sentence.lower().split()
    
    # Dictionary to store word counts
    word_counts = {}

    for word in words:
        # Using the .get() method as seen in your notes
        word_counts[word] = word_counts.get(word, 0) + 1

    # Finding the key with the maximum value
    most_repeated = max(word_counts, key=word_counts.get)
    
    print(f"The most repeated word is: '{most_repeated}'")
    print(f"It appears {word_counts[most_repeated]} times.")

# Example Usage:
text = "apple banana apple cherry banana apple"
most_repeated_word(text)
```

```python
#1. Swap Adjacent Elements
L = [11, 22, 33, 44, 55, 66]

# Iterate through the list, jumping 2 indices at a time
for i in range(0, len(L) - 1, 2):
    # Standard Pythonic swap
    L[i], L[i+1] = L[i+1], L[i]
print(L)
```

```python
#2.Reverse the Entire List
L = [11, 22, 33, 44, 55, 66]

# Set pointers for the start and end of the list
start = 0
end = len(L) - 1

while start < end:
    # Swap elements at start and end
    L[start], L[end] = L[end], L[start]
    
    # Move pointers toward the middle
    start += 1
    end -= 1

print(L)
```
